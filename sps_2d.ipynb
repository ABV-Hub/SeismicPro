```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfr = pd.read_csv('/home/egor/Desktop/2D_Valyton/sps/ALL_VALUNT0910_R_utm.txt', sep='\s+', header=None)
dfr.columns = ['rline', 'rid', 'x', 'y', 'z']
dfr.head()
```

```python
dfs = pd.read_csv('/home/egor/Desktop/2D_Valyton/sps/ALL_VALUNT0910_S_utm.txt', sep='\s+', header=None)
dfs.columns = ['sline', 'sid', 'point_index', 'sht_depth', 'uphole', 'x', 'y', 'z']
dfs = dfs.dropna(axis=0)
dfs.head()
```

```python
%matplotlib notebook
sline = 'S35'
rline = 'R9'
sid = 1011
rid = 151

plt.scatter(dfs['x'], dfs['y'], s=10, label="S")
plt.scatter(dfr['x'].values, dfr['y'].values, s=1, label="R")
for line in dfs['sline'].unique():
    p = dfs[(dfs['sline'] == line) & (dfs['sid'] == 1011)]
    plt.text(p['x'] - 2000, p['y'], line[1:])

plt.scatter(dfs[dfs['sline'] == sline]['x'], dfs[dfs['sline'] == sline]['y'], label="S_line", c='black', s=10)
plt.scatter(dfr[dfr['rline'] == rline]['x'], dfr[dfr['rline'] == rline]['y'], label="R_line", c='r', s=10)
plt.scatter(dfs[dfs['sid'] == sid]['x'], dfs[dfs['sid'] == sid]['y'], label="Sid", c='g')
plt.scatter(dfr[dfr['rid'] == rid]['x'], dfr[dfr['rid'] == rid]['y'], label="Rid", c='y', s=20)
plt.legend()
plt.show()
```

```python
dfx = pd.read_csv('/home/egor/Desktop/2D_Valyton/sps/ALL_VALUNT0910_X.sps',
                  sep='\s+', header=None, skiprows=49)
dfx.columns = ['type_num', 'xid', 'sline', 'sid', 'from_channel', 'to_channel', 'rline',
               'from_receiver', 'to_receiver']
dfx['sline'] = dfx['sline'].apply(lambda x: "S" + str(x))
dfx['rline'] = dfx['rline'].apply(lambda x: "R" + str(x))
dfx.head()
```
