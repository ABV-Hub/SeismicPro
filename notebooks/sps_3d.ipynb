```python
def add_delims(s, pos, delim):
    if type(pos) is int:
        pos = [pos]
    for i, p in enumerate(pos):
        aslist = list(s)
        aslist.insert(p + i * len(delim), delim)
        s = ''.join(aslist)
    return s

with open('/home/egor/Desktop/Xfield/Xfiled.Ssps', "r") as f:
    lines = f.readlines()

with open('/home/egor/Desktop/Xfield/Xfiled_fixed.Ssps', "w") as f:
    for line in lines:
        f.writelines(add_delims(line, [26, 40], ' '))
```

```python
dfx = pd.read_csv('/home/egor/Desktop/Xfield/Xfiled_fixed.Xsps',
                  sep='\s+', header=None, skiprows=0)
dfx.columns = ['type_num', 'xid', 'sline', 'sid', 'from_channel', 'to_channel', 'rline',
               'from_receiver', 'to_receiver']
dfx['sline'] = dfx['sline'].apply(lambda x: "S" + str(x))
dfx['rline'] = dfx['rline'].apply(lambda x: "R" + str(x))
dfx['from_receiver'] = dfx['from_receiver'].apply(lambda x: x * 10 + 1)
dfx['to_channel'] = dfx['to_channel'].apply(lambda x: int((x - 1) / 10))
dfx.head()
```

```python
%matplotlib notebook
segd = dfx[dfx['xid'] == '250921Z']

plt.scatter(dfs['x'], dfs['y'], s=10, label="S")
plt.scatter(dfr['x'].values, dfr['y'].values, s=1, label="R")

for index, row in segd.iterrows():
    view = dfr[(dfr['rline'] == row['rline']) &
               (dfr['rid'] >= row['from_receiver']) &
               (dfr['rid'] <= row['to_receiver'])]
    plt.scatter(view['x'], view['y'], label="Receivers", c='black', s=10)

    view = dfs[(dfs['sline'] == row['sline']) & (dfs['sid'] == row['sid'])]
    plt.scatter(view['x'], view['y'], label="Shot", c='r', s=20)

plt.show()
```
