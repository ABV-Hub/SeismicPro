## Загрузка по индексу

```python
batch = (Dataset(field_index, SeismicBatch).next_batch(1)
         .load(sort_by='trace_number', components=('raw', 'lift')))
```

Просмотр загруженных компонент

```python
(batch.to_2d(component='raw')
 .show_slice(batch.indices[0], component='raw', cmap='gray', vmin=-0.1, vmax=0.1))
```

```python
(batch.to_2d(component='lift')
 .show_slice(batch.indices[0], component='lift', cmap='gray', vmin=-0.1, vmax=0.1))
```

## Параллельная vs последовательная загрузка

Параллельная загрузка в несколько компонент обычно происходит быстрее последовательной загрузки в компоненты. Ускорение зависит от объема считываемых данных.

```python
%%timeit
(Dataset(trace_index, SeismicBatch).next_batch(100)
 .load(sort_by='trace_number', components=('raw', 'lift')))
```

```python
%%timeit
(Dataset(trace_index, SeismicBatch).next_batch(100)
 .load(sort_by='trace_number', components='raw')
 .load(sort_by='trace_number', components='lift'))
```

На больших объемах скорости загрузки становятся близкими

```python
%%timeit
(Dataset(trace_index, SeismicBatch).next_batch(10000)
 .load(sort_by='trace_number', components=('raw', 'lift')))
```

```python
%%timeit
(Dataset(trace_index, SeismicBatch).next_batch(10000)
 .load(sort_by='trace_number', components='raw')
 .load(sort_by='trace_number', components='lift'))
```
